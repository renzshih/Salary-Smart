```python
def Philhealth(x):
    if x <= 10000:
        Philhealth = 137.5
    elif x > 10000 and x <= 39999.99:
        Philhealth = x*0.0275/2
    else:
        Philhealth = 550
    return Philhealth
```

```python
def Pagibig(x):
    if x <= 1500:
        Pagibig = x*0.001
    else:
        Pagibig = x*0.002
    return Pagibig
```

```python
def IncomeTax(x):
    if x <= 250000/12:
        IncomeTax = 0
    elif x > 250000/12 and x <= 400000/12:
        IncomeTax = (x-250000/12)*0.15
    elif x > 400000/12 and x <= 800000:
        IncomeTax = 22500/12 + (x-400000/12)*0.20
    elif x > 800000/12 and x <= 2000000/12:
        IncomeTax = 102500/12 + (x-800000/12)*0.25
    elif x > 2000000/12 and x <= 8000000/12:
        IncomeTax = 402500/12 + (x-2000000/12)*0.30
    else:
        IncomeTax = 2202500/12 + (x-8000000/12)*0.35
    return IncomeTax
```

```python
# Dictionary Version for csv
import csv
with open("SSS.csv",newline="") as csv_file:
    SSS = csv.DictReader(csv_file)
    SSS_csv = []
    for i in SSS:
        Bottom_value_of_Range = i['Bottom value of Range']
        Highest_value_of_range = i['Highest value of range']
        Monthly_Salary_Credit = i['Monthly Salary Credit']
        Employee_Contribution = i['Employee Contribution']
        Employer_Contribution = i['Employer Contribution']
        SSS_csv.append([Bottom_value_of_Range,Highest_value_of_range,Monthly_Salary_Credit,Employee_Contribution,Employer_Contribution])
```
